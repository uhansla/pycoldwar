```python
from lupa import LuaRuntime
import pandas as pd
from dcs.weapons_data import Weapons
import dcs_util as dcs_util
from collections import defaultdict


# --- Step 2: Load and parse Lua mission ---

mission_data, lua = dcs_util.load_mission("Pretense_Caucasus_2.0.alpha4\\Pretense_Caucasus_2.0.alpha4\\mission")


# --- Step 5: Extract and display units ---
df = dcs_util.get_planes(mission_data)
df[df['Skill']!='Client']
```

```python
cold_mission, lua_cold = dcs_util.load_mission("coldwar\\mission")
cold_df = dcs_util.get_planes(cold_mission)
cold_df[cold_df['Skill']!='Client']
```

```python
blue_units = pd.unique(cold_df[(cold_df['Skill']!='Client') & (cold_df['Side']=='Blue')]['Unit Type'])
red_units = pd.unique(cold_df[(cold_df['Skill']!='Client') & (cold_df['Side']=='Red')]['Unit Type'])

mission_types = ['sead', 'patrol', 'strike', 'cas']

blue_map = defaultdict(list)
red_map = defaultdict(list)

# Loop over all combinations of unit_type and mission_type
for mission_type in mission_types:
    for unit_type in blue_units:
        try:
            sample = dcs_util.find_sample_plane_unit(cold_mission, 'blue', unit_type, mission_type)
            if sample:
                blue_map[mission_type].append(sample)
        except ValueError:
            continue  # skip if not found

    for unit_type in red_units:
        try:
            sample = dcs_util.find_sample_plane_unit(cold_mission, 'red', unit_type, mission_type)
            if sample:
                red_map[mission_type].append(sample)
        except ValueError:
            continue
```

```python
dcs_util.apply_sample_templates(mission_data, blue_map, red_map)

mod_df = dcs_util.get_planes(mission_data)
mod_df[mod_df['Skill']!='Client']

```

```python
ground_units = dcs_util.get_ground_units(mission_data)
ground_units
```

```python
cold_ground_units = dcs_util.get_ground_units(cold_mission)
cold_ground_units
```

```python
ground_map = dcs_util.extract_ground_templates(cold_mission)
```

```python
dcs_util.apply_ground_templates(mission_data, ground_map)
mod_ground_units = dcs_util.get_ground_units(mission_data)
mod_ground_units
```

```python
heli = dcs_util.get_helicopters(mission_data)
heli[heli['Skill']!='Client']
```

```python
cold_heli = dcs_util.get_helicopters(cold_mission)
cold_heli[cold_heli['Skill']!='Client']
```

```python
heli_map = dcs_util.extract_helicopter_templates(cold_mission)
dcs_util.apply_helicopter_templates(mission_data, heli_map)
heli = dcs_util.get_helicopters(mission_data)
heli[heli['Skill']!='Client']
```

```python
# Update Mission File
print(lua)
mission_modified_lua = dcs_util.dict_to_lua_table(mission_data, lua)
lua.globals().mission = mission_modified_lua

# Use this to generate the full output script
lua_script = f"mission = {dcs_util.serialize_lua_table(mission_data)}"

output_path = "output\\mission"

with open(output_path, "w", encoding="utf-8") as f:
    f.write(lua_script)

print(f"✅ Modified mission written to {output_path}")

```
